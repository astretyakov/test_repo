```python
import pandas as pd

df_log = pd.DataFrame(columns=["Номер модели",
                               "Объём словаря",
                               "Размер окна",
                               "Шаг разбиения",
                               "Результат точности"])

model_number = 1
vocabulary_size = 2
window_size = 3
window_step = 4
accuracy_result = 5

df_log.loc[len(df_log.index)] = [model_number, vocabulary_size,
                                 window_size, window_step, accuracy_result]

display(df_log)
```

```python
for i in range(2, 21):
  model_number = i
  vocabulary_size = i*2
  window_size = i*3
  window_step = i*4
  accuracy_result = i*5

  df_log.loc[len(df_log.index)] = [model_number, vocabulary_size,
                                   window_size, window_step, accuracy_result]

display(df_log)
```
